```python

df['Year'] = df['Date'].dt.year

# Number of Orders per Year
orders_per_year = df.groupby('Year').size()
print("\nOrders per Year:")
print(orders_per_year)

# Most Popular Type of Meal
popular_meal = df['Type of Meal'].value_counts()
print("\nMost Popular Meal Types:")
print(popular_meal)
```

```python
# Total Revenue
total_revenue = df['Meal Price'].sum()
print(f"\nTotal Revenue: {total_revenue}")

# Revenue Contribution by Meal Type
revenue_by_meal_type = df.groupby('Type of Meal')['Meal Price'].sum()
print("\nRevenue by Meal Type:")
print(revenue_by_meal_type)
```

```python
# Add a Count of Participants Column
df['Participants Count'] = df['Participants'].apply(len)

# Average Participants per Meal
avg_participants = df['Participants Count'].mean()
print(f"\nAverage Participants per Meal: {avg_participants}")

# Most Frequent Participants
participant_frequency = df.explode('Participants')['Participants'].value_counts()
print("\nMost Frequent Participants:")
print(participant_frequency)
```

```python
# Maximum and Minimum Meal Prices
max_price = df['Meal Price'].max()
min_price = df['Meal Price'].min()

# Meals with Highest and Lowest Prices
highest_priced_meal = df[df['Meal Price'] == max_price]
lowest_priced_meal = df[df['Meal Price'] == min_price]

print(f"\nHighest Priced Meal: {max_price}")
print(highest_priced_meal)

print(f"\nLowest Priced Meal: {min_price}")
print(lowest_priced_meal)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Orders per Year
orders_per_year.plot(kind='bar', color='skyblue', figsize=(8, 5))
plt.title('Number of Orders per Year')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Plot Revenue by Meal Type
revenue_by_meal_type.plot(kind='bar', color='teal', figsize=(8, 5))
plt.title('Total Revenue by Meal Type')
plt.xlabel('Type of Meal')
plt.ylabel('Revenue')
plt.show()
```

```python
# Plot Participant Frequency
participant_frequency.head(5).plot(kind='bar', color='coral', figsize=(8, 5))
plt.title('Top 5 Participants by Attendance')
plt.xlabel('Participant')
plt.ylabel('Number of Meals')
plt.show()
```
